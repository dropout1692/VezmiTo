```python
import pandas as pd
```

```python
fruit_and_veggies = pd.read_csv('./fruit_and_veggies.csv')
fruit_and_veggies.head()
```

```python
herbs = pd.read_csv('./herbs.csv')
herbs.head()
```

```python
mushrooms = pd.read_csv('./mushrooms.csv')
mushrooms.head()
```

```python
merged = pd.concat([fruit_and_veggies, herbs, mushrooms])
merged['id'] = merged['id'].str.capitalize()
len(merged)
```

```python
merged.head()
```

```python
merged[['id', 'start', 'end', 'edible', 'type']].to_json('./merged.json', orient = 'records', force_ascii=False)
```
