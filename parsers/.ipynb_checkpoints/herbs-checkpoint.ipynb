```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
with open("./herbs.html") as f:
    soup = BeautifulSoup(f, 'html.parser')
```

```python
herbs_list = soup.find('div', class_="herbs-list")
herbs = herbs_list.find_all('div', class_='herb')
# herbs = [herbs[0]]
len(herbs)
```

```python
month_index_map = {
 'január': 1,
 'február': 2,
 'marec': 3,
 'apríl': 4,
 'máj': 5,
 'jún': 6,
 'júl': 7,
 'august': 8,
 'september': 9,
 'október': 10,
 'november': 11,
 'december': 12
}

obj = {
    'id': [],
    'start': [],
    'end': []
}

for herb in herbs:
    try:
        title = herb.find("h3").text
        print('processing', title)
        href = herb.find('a')['href']
        r = requests.get(href)
        soup2 = BeautifulSoup(r.text, 'html.parser')
        headlines = soup2.find_all('h3')
        harvest_season_headline = [item for item in headlines if "doba zberu" in item.text.lower()][0]
        months = [item.strip().lower() for item in harvest_season_headline.text.split(":")[1].split(",")]
        month_indexes = [month_index_map[item] for item in months]
        obj['id'].append(title)
        obj['start'].append(min(month_indexes))
        obj['end'].append(max(month_indexes))
    except Exception as e:
        print('skipping', title, e)
```

```python
df = pd.DataFrame(obj)
df['id'] = df['id'].str.capitalize()
df['edible'] = 1
df['type'] = 2 # herb
df
```

```python
df.to_csv("./herbs.csv")
```
